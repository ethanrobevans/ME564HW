# 1-2 
A given mass x of a radioactive element obeys the following differential equation in time:
$$x' = \lambda*x$$

a) Write down the solution:
$$\boxed{x(t) = e^{\lambda*t}}$$


```python
t = np.linspace (0,5,1000)
xnaught = 2
lamb = [-5,-1,0,0.01,.1]
for i in lamb:
    x = xnaught*np.exp(i*t)
    plt.plot(t,x,label = 'Lambda = '+str(i))
plt.xlabel('time')
plt.ylabel('Radioactive Mass')
plt.legend();
```

c) The half-life T is defined as the time it takes for the material to be reduced to half of its mass
through radioactive decay. The half-life of uranium-238 is 4.468 billion years. What is the
corresponding value of λ?

If we start with 1 kg of material, then x(0) = 1 and x(t* = 4.468 billion years) = 1/2
Solving our solution for lambda: 
$$x(t*) = x(0)*e^{\lambda*t*}$$\
$$\lambda = ln(\frac{x(t*)}{x(0)})*\frac{1}{t*}$$

Thus $$\boxed{\lambda = -0.1551 (billion years)^{-1}}$$

d)  If you start with 100kg of uranium-238, how long until you only have 5kg left?
Using our Lambda Value and solving for time:
$$t* = ln(\frac{x(t*)}{x(0)})*\frac{1}{\lambda}$$

$$t = ln(\frac{5}{100})*\frac{1}{-0.1551}$$

$$\boxed{t* = 19.31 billion years}$$




# 1-3 Taylor Series
Compute the taylor series and graph the first 3 nonzero terms for each of the following

## a)
$$f(x) = \frac{sin(x)}{x}$$

To simplify, we will take the taylor series of sin(x) and then divide by x:
$$g(x) = sin(x)$$
$$g'(x) = cos(x)$$
$$g''(x) = -sin(x)$$
$$g'''(x) = -cos(x)$$
$$g''''(x) = sin(x)$$

The Taylor Series of g(x) around a=1 is now:
$$g(x) = sin(1) + cos(1)(x-1)-\frac{sin(1)}{2!}(x-1)^2-\frac{cos(1)}{3!}(x-1)^3+...$$

Now we divide by x to get f(x) (Note that this new function has an explosion at 0):
$$\boxed{f(x) =  \frac{sin(1)}{x} + \frac{cos(1)}{x}(x-1)-\frac{sin(1)}{2!x}(x-1)^2-\frac{cos(1)}{3!x}(x-1)^3+...}$$
Now to plot...

```python
x = np.linspace(-5,5,1000)
y1 = np.sin(x)/x
plt.plot(x,y1,label = 'sin(x)/x')
s = np.sin(1)
c = np.cos(1)
y2num = s+c*(x-1)-s/2*(x-1)**2
y2 = y2num/x
plt.plot(x,y2, label = 'Taylor Series 3 Term')
y3num = s+c*(x-1)-s/2*(x-1)**2-c/6*(x-1)**3+s/(4*3*2)*(x-1)**4+c/(5*4*3*2)*(x-1)**5
y3 = y3num/x
plt.plot(x,y3, label = 'Taylor Series 6 Term')
plt.xlabel('x')
plt.ylabel('y')
plt.ylim(-1.5,1.5)
plt.legend();
```

```python
x = np.linspace(-5,5,1000)
err1 = y2-y1
err2 = y3-y1
plt.plot(x,err1, label = 'Error with 3 terms')
plt.plot(x,err2, label = 'Error with 6 terms')
plt.axhline(y = 0, c = 'k', linewidth = 1)
plt.legend();
```

## b)

$$f(x) = 3^{x}$$

First we will take some derivitives:
$$f'(x) = ln(3)*3^{x}$$
$$f''(x) = ln(3)^{2}*3^{x}$$
$$f'''(x) = ln(3)^{3}*3^{x}$$

We can clearly see:
$$f^n(x) = ln(3)^n*3^x$$
And evaluating those derivitives at zero gives:
$$f^n(0) = ln(3)^n$$


Plugging this into the Taylor Series Definition with a=0 clearly gives the first few terms as:
$$f(x) = 1+ln(3)*x+\frac{ln(3)^2}{2!}*x^2+\frac{ln(3)^3}{3!}*x^3+...$$


Writing the above as a summation:
$$\boxed{f(x) = \sum_{n=0}^{\infty} \frac{ln(3)^n*x^n}{n!}}$$
Now to plot...

```python
x = np.linspace(-5,5,1000)
y1 = 3**x
plt.plot(x,y1,label = '3^x')
ln3 = np.log(3)
y2 = 1 + ln3*x + ln3**2*x**2/2
plt.plot(x,y2, label = 'Taylor Series 3 Term')
y3 = 1 + ln3*x + ln3**2*x**2/2 + ln3**3*x**3/6 + ln3**4*x**4/(4*3*2) + ln3**5*x**5/(5*4*3*2)
plt.plot(x,y3, label = 'Taylor Series 6 Term')
plt.xlabel('x')
plt.ylabel('y')
plt.legend();
```

```python
x = np.linspace(-5,5,1000)
err1 = y2-y1
err2 = y3-y1
plt.plot(x,err1, label = 'Error with 3 terms')
plt.plot(x,err2, label = 'Error with 6 terms')
plt.axhline(y = 0, c = 'k', linewidth = 1)
plt.legend();
```

# 1-4 Complex Expressions

Compute the analytic expression for each of the following and plot them


## a)
Using Eulers Formula, we get:
$$\boxed{e^{i*t} = cos(t) +i*sin(t)}$$

Real Part: $$cos(t)$$
Imaginary Part: $$sin(t)$$

```python
t = np.linspace(0,10,1000)
y = np.sin(t) #Imaginary Part
x = np.cos(t) #Real Part
plt.plot(x,y)
plt.xlabel('Real')
plt.ylabel('Imaginary')
plt.axhline(y = 0, c = 'k', linewidth = 1)
plt.axvline(x = 0, c = 'k', linewidth = 1);
```

## b) 
$$e^{(-1-i)*t} = e^{-t}e^{(i*t)^{-1}}$$
Plugging in Eulers, we get:
$$e^{-t}(cos(t)+isin(t))^{-1} = \boxed{e^{-t}cos(-t)+ie^{-t}sin(-t)}$$

Real Part: $$e^{-t}cos(-t)$$
Imaginary Part: $$e^{-t}sin(-t)$$

```python
t = np.linspace(0,10,1000)
y = np.exp(-t)*np.sin(-t) #Imaginary Part
x = np.exp(-t)*np.cos(-t) #Real Part
plt.plot(x,y)
plt.xlabel('Real')
plt.ylabel('Imaginary')
plt.axhline(y = 0, c = 'k', linewidth = 1)
plt.axvline(x = 0, c = 'k', linewidth = 1);
```

## c)
$$e^{1-it} = e^{1}*e{-it}$$

$$\boxed{e^{1}cos(-t)+e^{1}sin(-t)}$$

Real Part: $$e^{1}cos(-t)$$
Imaginary Part: $$e^{1}sin(-t)$$

```python
t = np.linspace(0,10,1000)
y = np.exp(1)*np.sin(-t) #Imaginary Part
x = np.exp(1)*np.cos(-t) #Real Part
plt.plot(x,y)
plt.xlabel('Real')
plt.ylabel('Imaginary')
plt.axhline(y = 0, c = 'k', linewidth = 1)
plt.axvline(x = 0, c = 'k', linewidth = 1);
```

## d)

$$e^{(-0.2+3\pi*i)t} = e^{-0.2t}*e{3\pi*it}$$

$$\boxed{e^{-0.2t}cos(3\pi*t)+e^{-0.2t}sin(3\pi*t)}$$

Real Part: $$e^{-0.2t}cos(3\pi*t)$$
Imaginary Part: $$e^{-0.2t}sin(3\pi*t)$$

```python
t = np.linspace(0,10,1000)
y = np.exp(-0.2*t)*np.sin(3*np.pi*t) #Imaginary Part
x = np.exp(-0.2*t)*np.cos(3*np.pi*t) #Real Part
plt.plot(x,y)
plt.xlabel('Real')
plt.ylabel('Imaginary')
plt.axhline(y = 0, c = 'k', linewidth = 1)
plt.axvline(x = 0, c = 'k', linewidth = 1);
```
